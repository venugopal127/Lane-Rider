```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values (time points)
x = np.linspace(0, 2 * np.pi, 100)  # Creating 1000 evenly spaced points from 0 to 2*pi

# Calculate sine values for each x value
y = np.sin(x)

# Create the plot
plt.figure(figsize=(8, 6))  # Optional: set the figure size
plt.scatter(x, y)

# Add labels and title
plt.xlabel('Time')
plt.ylabel('Amplitude')


# Show the plot
plt.show()
```
